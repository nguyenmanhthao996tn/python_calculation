```python
# ===== Include all needed libraries / packages =====


import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ===== Create list of frequencies & dataframes from csv files =====


# Specify the directory where the CSV files are
directory = './csv_data/'

# Specify a list of all CSV filenames in the directory
csv_files = 'ps150modif - Copy.csv'

# Create a list of DataFrames
df = pd.read_csv(os.path.join(directory, csv_files))
```

```python
# ===== Plot data =====

min_plot_freq = (140 - 50)
max_plot_freq = (140 + 50)
step_plot_freq = 10

%matplotlib inline

fig, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(6, 4), layout='constrained')

filtered_df = df[df['freq[MHz]'].between(min_plot_freq, max_plot_freq, inclusive="neither")]

list_freq = filtered_df['freq[MHz]'].values.flatten()

# Phase
# Phase - Port 1
y_axis_data = filtered_df['ang:Trc1_S21'].values.flatten()
ax1.plot(list_freq, y_axis_data, label='Phase Input/Port 1')

# Phase - Port 2
y_axis_data = filtered_df['ang:Trc2_S31'].values.flatten()
ax1.plot(list_freq, y_axis_data, label='Phase Input/Port 2')

# Phase - Port 3
y_axis_data = filtered_df['ang:Trc3_S41'].values.flatten()
ax1.plot(list_freq, y_axis_data, label='Phase Input/Port 3')

# Phase - Port 4
y_axis_data = filtered_df['ang:Mem5[Trc3]_S41'].values.flatten()
ax1.plot(list_freq, y_axis_data, label='Phase Input/Port 4')

# Annotate
freq_annotate = 140

annotate_text_x_offset = 1
annotate_text_y_offset = 10

plt.axvline(x=freq_annotate, color='r', linestyle='--')

# Annotate - Port 1
y_value = filtered_df[filtered_df['freq[MHz]'] == freq_annotate]['ang:Trc1_S21'].iloc[0]
plt.plot(freq_annotate, y_value, marker='o', color='k')
plt.text(freq_annotate + annotate_text_x_offset, y_value + annotate_text_y_offset, "{:.3f}".format(y_value))

# Annotate - Port 2
y_value = filtered_df[filtered_df['freq[MHz]'] == freq_annotate]['ang:Trc2_S31'].iloc[0]
plt.plot(freq_annotate, y_value, marker='o', color='k')
plt.text(freq_annotate + annotate_text_x_offset, y_value + annotate_text_y_offset, "{:.3f}".format(y_value))

# Annotate - Port 3
y_value = filtered_df[filtered_df['freq[MHz]'] == freq_annotate]['ang:Trc3_S41'].iloc[0]
plt.plot(freq_annotate, y_value, marker='o', color='k')
plt.text(freq_annotate + annotate_text_x_offset, y_value + annotate_text_y_offset, "{:.3f}".format(y_value))

# Annotate - Port 4
y_value = filtered_df[filtered_df['freq[MHz]'] == freq_annotate]['ang:Mem5[Trc3]_S41'].iloc[0]
plt.plot(freq_annotate, y_value, marker='o', color='k')
plt.text(freq_annotate + annotate_text_x_offset, y_value + annotate_text_y_offset, "{:.3f}".format(y_value))

# Axis configuration
ax1.set_xticks(np.arange(min_plot_freq, max_plot_freq + step_plot_freq, step_plot_freq))
ax1.set_yticks(np.arange(-180, 181, 30))
ax1.set_xlabel('Frequency (MHz)')
ax1.set_ylabel('Phase (deg)')
# ax1.set_title("Return Loss")
ax1.tick_params(axis='y')
ax1.grid(True)
ax1.legend()

plt.savefig(("ps140_phase.pdf"), format='pdf', dpi=600)

plt.show()
```

```python
# ===== Plot data =====

min_plot_freq = (140 - 50)
max_plot_freq = (140 + 50)
step_plot_freq = 10

%matplotlib inline

fig, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(6, 4), layout='constrained')

filtered_df = df[df['freq[MHz]'].between(min_plot_freq, max_plot_freq, inclusive="neither")]

list_freq = filtered_df['freq[MHz]'].values.flatten()

# Phase
# Phase - Port 1
y_axis_data = filtered_df['db:Trc1_S21'].values.flatten()
ax1.plot(list_freq, y_axis_data, label='Input/Port 1')

# Phase - Port 2
y_axis_data = filtered_df['db:Trc2_S31'].values.flatten()
ax1.plot(list_freq, y_axis_data, label='Input/Port 2')

# Phase - Port 3
y_axis_data = filtered_df['db:Trc3_S41'].values.flatten()
ax1.plot(list_freq, y_axis_data, label='Input/Port 3')

# Phase - Port 4
y_axis_data = filtered_df['db:Mem5[Trc3]_S41'].values.flatten()
ax1.plot(list_freq, y_axis_data, label='Input/Port 4')

# Total
filtered_df['p1_lin'] =  10 ** (filtered_df['db:Trc1_S21'] / 10)
filtered_df['p2_lin'] =  10 ** (filtered_df['db:Trc2_S31'] / 10)
filtered_df['p3_lin'] =  10 ** (filtered_df['db:Trc3_S41'] / 10)
filtered_df['p4_lin'] =  10 ** (filtered_df['db:Mem5[Trc3]_S41'] / 10)
filtered_df['total_lin'] = filtered_df['p1_lin'] + filtered_df['p2_lin'] + filtered_df['p3_lin'] + filtered_df['p4_lin']
filtered_df['total'] = 10 * np.log10(filtered_df['total_lin'])

y_axis_data = filtered_df['total'].values.flatten()
ax1.plot(list_freq, y_axis_data, label='Insertion Loss')

# Annotate
freq_annotate = 140

# # annotate_text_x_offset = 0
# # annotate_text_y_offset = 0

plt.axvline(x=freq_annotate, color='r', linestyle='--')

# # Annotate - Port 1
y_value_1 = filtered_df[filtered_df['freq[MHz]'] == freq_annotate]['db:Trc1_S21'].iloc[0]
plt.plot(freq_annotate, y_value_1, marker='o', color='k')
# plt.text(freq_annotate + annotate_text_x_offset, y_value + annotate_text_y_offset, "{:.3f}".format(y_value))

# # Annotate - Port 2
y_value_2 = filtered_df[filtered_df['freq[MHz]'] == freq_annotate]['db:Trc2_S31'].iloc[0]
plt.plot(freq_annotate, y_value_2, marker='o', color='k')
# plt.text(freq_annotate + annotate_text_x_offset, y_value + annotate_text_y_offset, "{:.3f}".format(y_value))

# # Annotate - Port 3
y_value_3 = filtered_df[filtered_df['freq[MHz]'] == freq_annotate]['db:Trc3_S41'].iloc[0]
plt.plot(freq_annotate, y_value_3, marker='o', color='k')
# plt.text(freq_annotate + annotate_text_x_offset, y_value + annotate_text_y_offset, "{:.3f}".format(y_value))

# # Annotate - Port 4
y_value_4 = filtered_df[filtered_df['freq[MHz]'] == freq_annotate]['db:Mem5[Trc3]_S41'].iloc[0]
plt.plot(freq_annotate, y_value_4, marker='o', color='k')
# plt.text(freq_annotate + annotate_text_x_offset, y_value + annotate_text_y_offset, "{:.3f}".format(y_value))

y_value_total = filtered_df[filtered_df['freq[MHz]'] == freq_annotate]['total'].iloc[0]
plt.plot(freq_annotate, y_value_total, marker='o', color='k')

annotate_text = ""
annotate_text += "Input / P1 = "
annotate_text += "{:.3f}".format(y_value_1)
annotate_text += " dB\n"
annotate_text += "Input / P2 = "
annotate_text += "{:.3f}".format(y_value_2)
annotate_text += " dB\n"
annotate_text += "Input / P3 = "
annotate_text += "{:.3f}".format(y_value_3)
annotate_text += " dB\n"
annotate_text += "Input / P4 = "
annotate_text += "{:.3f}".format(y_value_4)
annotate_text += " dB\n\n"
annotate_text += "Total Insertion Loss = "
annotate_text += "{:.3f}".format(y_value_total)
annotate_text += " dB\n\n"


plt.text(freq_annotate + 7, -30, annotate_text)

# Axis configuration
ax1.set_xticks(np.arange(min_plot_freq, max_plot_freq + step_plot_freq, step_plot_freq))
ax1.set_yticks(np.arange(-30, 5, 5))
ax1.set_xlabel('Frequency (MHz)')
ax1.set_ylabel('dB')
ax1.set_title("Phase Shifter Insertion Loss")
ax1.tick_params(axis='y')
ax1.grid(True)
ax1.legend()

plt.savefig(("ps140_mag.pdf"), format='pdf', dpi=600)

plt.show()
```

```python
# Calculate output divided ratio

p1_ratio = (100 * 10 ** (filtered_df['db:Trc1_S21'] / 10)).values.flatten()
p2_ratio = (100 * 10 ** (filtered_df['db:Trc2_S31'] / 10)).values.flatten()
p3_ratio = (100 * 10 ** (filtered_df['db:Trc3_S41'] / 10)).values.flatten()
p4_ratio = (100 * 10 ** (filtered_df['db:Mem5[Trc3]_S41'] / 10)).values.flatten()

total_ratio = p1_ratio + p2_ratio + p3_ratio + p4_ratio;

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 4), layout='constrained')
ax.plot(list_freq, p1_ratio, label='Input/Port 1')
ax.plot(list_freq, p2_ratio, label='Input/Port 2')
ax.plot(list_freq, p3_ratio, label='Input/Port 3')
ax.plot(list_freq, p4_ratio, label='Input/Port 4')
ax.plot(list_freq, total_ratio, label='Total')

# Axis configuration
ax.set_xticks(np.arange(min_plot_freq, max_plot_freq + step_plot_freq, step_plot_freq))
ax.set_yticks(np.linspace(0, 100, 11))
ax.set_xlabel('Frequency (MHz)')
ax.set_ylabel('%')
ax.tick_params(axis='y')
ax.grid(True)
ax.legend()


# === Print to console
text = ""
text += 'Avg. Port 1: ' + "{:.3f}".format(np.mean(p1_ratio)) + "%\n"
text += 'Avg. Port 2: ' + "{:.3f}".format(np.mean(p2_ratio)) + "%\n"
text += 'Avg. Port 3: ' + "{:.3f}".format(np.mean(p3_ratio)) + "%\n"
text += 'Avg. Port 4: ' + "{:.3f}".format(np.mean(p4_ratio)) + "%\n"
text += 'Total: ' + "{:.3f}".format(np.mean(total_ratio)) + "%"
plt.text(100, 70, text)

plt.show()
```
